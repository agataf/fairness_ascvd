```python
import pandas as pd
import numpy as np

cohort_file= '/labs/shahlab/projects/agataf/bmj_paper/pce_data/all_cohorts.csv'

df = pd.read_csv(cohort_file)#.query("censored_10yr==False")

df_new = df.assign(grp = np.where((df.grp==1) & (df.race_black==0),
                              2,
                              np.where((df.grp==3) & (df.race_black==0),
                                        4,
                                        df.grp)
                             )
              )

df.iloc[np.where(~(df==df_new))[0]]
```
