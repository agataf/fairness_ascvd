```python
def read_file(filename, path, files_dict, index):
    df = pd.read_csv(os.path.join(path, '.'.join((filename, 'csv'))), engine='python')
    df = (df
          .rename(columns = lambda x: x.lower())
          .filter(files_dict[filename])
          .set_index(index)
          .sort_index()
         )

    return df

def read_file_events(filename, path, index):
    df = pd.read_csv(os.path.join(path, '.'.join((filename, 'csv'))), engine='python')
    df = (df
          .rename(columns = lambda x: x.lower())
          #.filter(files_dict[filename])
          .set_index(index)
          .sort_index()
         )

    return df
```

```python
def get_smokstat(df):
    smokstat = np.where((df.a10cigs==1), 3, np.nan)
    smokstat = np.where((df.a10cigs==2) & (df.a09smknw == 1), 2, smokstat)
    smokstat = np.where((df.a10cigs==2) & (df.a09smknw == 2), 1, smokstat)
    return smokstat

def get_cursmoke(df):
    cursmoke = np.where((df.smokstat==1), 1, np.nan)
    cursmoke = np.where(((df.smokstat==2) | (df.smokstat==3)), 0, cursmoke)
    return cursmoke

def get_hyptmdsr(df):
    hyptmdsr = np.where((df.a08bpmed == 1), 0, np.nan)
    hyptmdsr = np.where((df.a08bpmed == 2) & (df.a09mdnow == 2), 1, hyptmdsr)
    hyptmdsr = np.where((df.a08bpmed == 2) & (df.a09mdnow == 1), 0, hyptmdsr)
    hyptmdsr = np.where((df.a08bpmed == 8), np.nan, hyptmdsr)
    return hyptmdsr

def get_diabhx(df):
    diabhx = np.where((df.a08diab==2) & ((df.a09dibst>=3) | (df.a09dibst) <= 4), 1, np.nan)
    diabhx = np.where((df.a08diab==1) | ((df.a08diab==2) & (df.a09dibst) == 5), 0, diabhx)
    return diabhx

def get_diabt126(df):
    fast08 = np.where((df.a05fstmn/60 < 8), 0, np.nan)
    fast08 = np.where((df.a05fstmn/60 >= 8), 1, fast08)
    fast12 = np.where((df.a05fstmn/60 < 12), 0, np.nan)
    fast12 = np.where((df.a05fstmn/60 >= 12), 1, fast12)
    diabt126 = np.where((df.glucose >= 200) |
                        ((df.glucose >= 126) & (fast08>=1)) |
                        (df.diabhx == 1),
                        1, np.nan)
    diabt126 = np.where((df.glucose > 0) & (df.glucose < 126) &
                        (df.diabhx != 1),
                        0, diabt126)
    diabt126 = np.where((df.glucose >=126) & (df.glucose < 200) &
                        (fast08 <=0) & (df.diabhx == np.nan),
                        np.nan, diabt126)
    return diabt126


    
def get_prevchd(df):
    prevchd = np.where((df.a09hrtak==1) | (df.a08heart==1), 0, np.nan)
    prevchd = np.where(df.a09hrtak==2, 1, prevchd)
    prevchd = np.where(df.a08heart==8, np.nan, prevchd)
    return prevchd


def get_prevap(df):
    prevap = np.where((df.a09angin==1) | (df.a08heart==1), 0, np.nan)
    prevap = np.where(df.a09angin==2, 1, prevap)
    prevap = np.where(df.a08heart==8, np.nan, prevap)
    return prevap
    
```

```python
df_yr0 = None
for filename in yr0_file_dict.keys():
    df = pd.read_csv(os.path.join(dirname_yr0,filename))
    df = df.filter(yr0_file_dict[filename])
    df = df.set_index('PID')
    if df_yr0 is not None:
        df_yr0 = df_yr0.join(df)
    else:
        df_yr0 = df
        
df_yr15 = None
for filename in yr15_file_dict.keys():
    df = pd.read_csv(os.path.join(dirname_yr15,filename))
    df = df.filter(yr15_file_dict[filename])
    df = df.rename(columns = {'PID': 'pid'})
    df = df.set_index('pid')
    if df_yr15 is not None:
        df_yr15 = df_yr15.join(df)
    else:
        df_yr15 = df
        
### year 0 ###
prevchd = np.where((df_yr0.A09HRTAK==1) | (df_yr0.A08HEART==1), 0, np.nan)
prevchd = np.where((df_yr0.A09HRTAK==2), 1, prevchd)
prevchd = np.where((df_yr0.A09HRTAK==8), np.nan, prevchd)

prevap = np.where((df_yr0.A09ANGIN==1) | (df_yr0.A08HEART==1), 0, float("NaN"))
prevap = np.where((df_yr0.A09ANGIN==2), 1, prevap)
prevap = np.where((df_yr0.A09ANGIN==8), float("NaN"), prevap)

diabhx = np.where((df_yr0.A08DIAB==2) & ((df_yr0.A09DIBST>=3) | (df_yr0.A09DIBST) <= 4), 1, float("NaN"))
diabhx = np.where((df_yr0.A08DIAB==1) | ((df_yr0.A08DIAB==2) & (df_yr0.A09DIBST) == 5), 0, diabhx)

df_yr0 = df_yr0.assign(prevchd = prevchd, 
                       prevap = prevap, 
                       diabhx = diabhx,
                       exam = 0)

df_yr0 = df_yr0.rename(columns={'diabhx': 'diabhx1',
                       'prevap': 'prevap1',
                       'prevchd': 'prevchd1',
                       'A01AGE1': 'age1',
                       'A01RACE1': 'race1'})

### year 15 ###
smokstat = np.where((df_yr15.F10CIGS==1), 3, float("NaN"))
smokstat = np.where((df_yr15.F10CIGS==2) & (df_yr15.F09SMKNW == 1), 2, smokstat)
smokstat = np.where((df_yr15.F10CIGS==2) & (df_yr15.F09SMKNW == 2), 1, smokstat)

cursmoke = np.where((smokstat==1), 1, float("NaN"))
cursmoke = np.where((smokstat==2) | (smokstat == 3), 0, cursmoke)

hyptmdsr = np.where((df_yr15.F08HBNOW==1) | ((df_yr15.F08HBNOW == 2) & (df_yr15.F08HBP ==1)), 0, np.nan)
hyptmdsr = np.where((df_yr15.F08HBNOW==1) | ((df_yr15.F08HBNOW == 2) & (df_yr15.F08HBP ==2)), 1, hyptmdsr)
hyptmdsr = np.where((df_yr15.F08HBNOW==8), np.nan, hyptmdsr)

cholmed = np.where((df_yr15.F08HBNOW==1), 0, 1)
cholmed = np.where((df_yr15.F08HBNOW==np.nan), np.nan, cholmed) # issue: =np.nan isn't showing up but 'nan's exist

fast08 = np.where((df_yr15.F05FSTMN/60 < 8), 0, 1)
fast12 = np.where((df_yr15.F05FSTMN/60 < 12), 0, 1)
glucose = df_yr15.FL7GLU*0.94359772+6.979619035

diabt126 = np.where((glucose >=200) | ((glucose>=126) & (fast08>=1)), 1, np.nan)
diabt126 = np.where((glucose > 0) & (glucose < 126), 0, diabt126)

df_yr15 = df_yr15.assign(smokstat = smokstat, 
                         cursmoke = cursmoke, 
                         hyptmdsr = hyptmdsr,
                         weight = df_yr15.F20WGT*0.45,
                         bmi = df_yr15.F20BMI,
                         diabt126 = diabt126,
                         age = df_yr15.EX6_AGE,
                         exmdate = df_yr15.F02DATE,
                         sysbp = df_yr15.F02AVGSY,
                         diabp = df_yr15.F02AVGDI,
                         heartrte = df_yr15.F02PULSE*2)
```
