```python
import pandas as pd
old='/labs/shahlab/projects/agataf/data/pooled_cohorts/cohort_extraction/all_cohorts_test_eval.csv'
df = pd.read_csv(old)
```

```python
nums = df.groupby('grp').count().person_id.values

print(str(sum(nums)) + ' individuals: ' +  str(nums[0]) + ' Black women, ' 
      + str(nums[1]) + ' White women, ' 
      + str(nums[2]) + ' Black men and ' 
      + str(nums[3]) + ' White men ')
```

```python
numbers = df.groupby(['study', 'grp']).count().person_id

means = (df
         .groupby(['grp'])
         .mean()
         .filter(['age', 'ascvd_10yr', 'censored_10yr', 'event_time_10yr'])
         .round(4)
         .reset_index()
        )

means_all = (df
             .assign(grp='all')
             .groupby(['grp'])
             .mean()
             .filter(['age', 'ascvd_10yr', 'censored_10yr', 'event_time_10yr'])
             .round(3)
             .reset_index()
            )


means = pd.concat([means, means_all]).reset_index()#.set_index(['grp']).sort_values(by='study')

counts = (df
          .groupby('grp')
          .count()
          .filter(['person_id'])
          .rename(columns={'person_id': 'n'})
          .reset_index()
          .append({'n': df.count().person_id, 'grp': 'all'}, ignore_index=True)
          .reset_index()
         )

(means.assign(counts=counts.n).filter(['grp', 'counts', 'age', 'ascvd_10yr', 'censored_10yr'])
 .rename(columns={'grp': 'group'})
 .assign(group = lambda x: x.group.map({1: 'Black women', 2: 'White women',
                                       3: 'Black men', 4: 'White men', 'all': 'all'}))
)
```

```python
numbers = df.groupby(['study', 'grp']).count().person_id

means = (df
         .groupby(['grp', 'study'])
         .mean()
         .filter(['age', 'ascvd_10yr', 'censored_10yr', 'event_time_10yr'])
         .round(3)
         .reset_index()
        )

means_all = (df
             .assign(grp='all')
             .groupby(['grp','study'])
             .mean()
             .filter(['age', 'ascvd_10yr', 'censored_10yr', 'event_time_10yr'])
             .round(3)
             .reset_index()
            )


means = pd.concat([means, means_all]).reset_index()#.set_index(['grp']).sort_values(by='study')

counts = (df
          .groupby(['study', 'grp'])
          .count()
          .filter(['person_id'])
          .rename(columns={'person_id': 'n'})
          .reset_index()
          .append({'n': df.count().person_id, 'grp': 'all'}, ignore_index=True)
          .reset_index()
         )

means.assign(counts=counts.n).set_index(['grp', 'study', ]).filter(['counts', 'age', 'ascvd_10yr', 'censored_10yr'])
```
