```python
aggregate_path_all = '/labs/shahlab/projects/agataf/data/cohorts/pooled_cohorts/experiments/bmj_manuscript/'

plot_df_all = pd.read_csv(os.path.join(aggregate_path_all, 'bootstrap_standard_eval_new.csv'))
plot_df = plot_df_all
```

```python
plot_df = plot_df_all.assign(model_type = lambda x: pd.Categorical(x.model_type, 
                                               categories = ['PCE', 'rPCE', 'BL', 'rBL', 'EO1', 'EO2', 'EO3', 'EO4'],
                                               ordered=True)
                  )
```

```python
grouped_plot_df = (plot_df
                   .query("metric==['sensitivity', 'specificity', 'impl_threshold']")
                   .drop(columns=['phase', 'CI_lower', 'CI_upper'])
                   .groupby(['model_type', 'thresholds', 'metric', 'group']).sum()
                   .reset_index()
                   .pivot(index=['model_type', 'thresholds', 'metric'], columns='group', values='CI_med')
                   .drop(columns=['overall'])
                  )
```

```python
means = grouped_plot_df.apply(np.mean, axis=1)
stds = grouped_plot_df.apply(np.std, axis=1)
```

```python
coeff_var = ((stds/means)
             .to_frame()
             .reset_index()
             .pivot(index=['thresholds','model_type'], columns=['metric'], values=0)
             .reset_index()
             .set_index(['thresholds', 'model_type', 'impl_threshold'])
             .stack()
             .to_frame()
             .reset_index()
             .rename(columns = {0: 'Inter group coefficient of variation',
                               'impl_threshold': 'IGCV in implied threshold',
                               'model_type': 'Model',
                               'metric': 'Metric',
                               'thresholds': 't'})
             .assign(t = lambda x: x.t.map({0.075: '7.5%', 0.2: '20%'}),
                     Metric = lambda x: x.Metric.map({'sensitivity': 'IGCV in sensitivity',
                                                      'specificity': 'IGCV in specificity'}
                                                    )
                    )
            )



coeff_var.head()
```

```python
std_frame = (stds
             .to_frame()
             .reset_index()
             .pivot(index=['thresholds','model_type'], columns=['metric'], values=0)
             .reset_index()
             .set_index(['thresholds', 'model_type', 'impl_threshold'])
             .stack()
             .to_frame()
             .reset_index()
             .rename(columns = {0: 'Inter group standard deviation',
                               'impl_threshold': 'IGSD in implied threshold',
                               'model_type': 'Model',
                               'metric': 'Metric',
                               'thresholds': 't'})
             .assign(t = lambda x: x.t.map({0.075: '7.5%', 0.2: '20%'}),
                     Metric = lambda x: x.Metric.map({'sensitivity': 'IGSD in sensitivity',
                                                      'specificity': 'IGSD in specificity'}
                                                    )
                    )
            )



std_frame.head()
```

```python
custom_palette = ["red", "hotpink", "darkgreen", "mediumseagreen", "lightskyblue", "deepskyblue", "blue", "black"]
sns.set_palette(custom_palette)

g=sns.relplot(data = coeff_var,
                x = 'IGCV in implied threshold',
                y = 'Inter group coefficient of variation',
                col='t',
                kind = 'scatter',
                hue='Model',
                style='Metric',
                facet_kws= {'sharey': False, 'sharex': False, 'margin_titles': True},
                aspect=1,
                s=80,

             )
```
